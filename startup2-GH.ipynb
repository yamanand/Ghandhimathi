```python
import statsmodels.formula.api as smf # for regression model
         
ml1 = smf.ols('Profit~ RD+Admin+Marketing', data=startup).fit() # regression model

# Summary
ml1.summary()
```

```python
# Finding rsquared values
ml1.rsquared , ml1.rsquared_adj  # Model accuracy is 94.75%
```

```python
# calculating VIF's values of independent variables
rsq_RD= smf.ols('RD ~ Admin + Marketing', data=startup).fit().rsquared  
vif_RD= 1/(1-rsq_RD) 
vif_RD
```

```python
rsq_Admin= smf.ols('Admin~ RD+Marketing+State', data=startup).fit().rsquared  
vif_Admin = 1/(1-rsq_Admin)
vif_Admin
```

```python
rsq_Marketing = smf.ols('Marketing ~ RD+Admin+State', data=startup).fit().rsquared  
vif_Marketing = 1/(1-rsq_Marketing) 
vif_Marketing
```

```python
# Storing vif values in a data frame
d1 = {'Variables':['RD','Admin','Marketing'],'VIF':[vif_RD,vif_Admin,vif_Marketing]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

```python
### Splitting the data into train and test data 
from sklearn.model_selection import train_test_split
startup_train, startup_test  = train_test_split(startup, test_size = 0.3) # 30% test data

# preparing the model on train data 
ml1_train = smf.ols("Profit ~ RD+Admin+Marketing", data = startup_train).fit()
# prediction on test data set 
test_pred = ml1_train.predict(startup_test)
# test residual values 
test_resid  = test_pred - startup_test.Profit
# RMSE value for test data 
test_rmse = np.sqrt(np.mean(test_resid*test_resid)) 
test_rmse

```

```python
# train_data prediction
train_pred = ml1_train.predict(startup_train)

# train residual values 
train_resid  = train_pred - startup_train.Profit

# RMSE value for train data 
train_rmse = np.sqrt(np.mean(train_resid * train_resid))
train_rmse
```

```python
# Test for Homoscedasticity or Heteroscedasticity (plotting model's standardized fitted values vs standardized residual values)
# User defined z = (x - mu)/sigma
def standard_values(vals) : return (vals-vals.mean())/vals.std()
plt.scatter(standard_values(ml1.fittedvalues),standard_values(ml1.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show()
```

```python
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(ml1)
plt.show()
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
startup2=startup.drop(startup.index[[49]],axis=0).reset_index(drop=True)
startup2
```

```python
ml2=smf.ols("Profit~RD+Admin+Marketing",data=startup2).fit()
ml2.summary()
```

```python
# 1. Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
(c,_)=ml2.get_influence().cooks_distance
c
```

```python
while ml2.rsquared < 0.99:
    for c in [np.max(c)>1]:
        ml2=smf.ols("Profit ~ RD + Admin + Marketing",data=startup2).fit()
        (c,_)=ml2.get_influence().cooks_distance
        c
        np.argmax(c) ,np.max(c)
        startup2=startup2.drop(startup2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        startup2
    else:
        final_model=smf.ols("Profit ~ RD + Admin + Marketing",data=startup2).fit()
        final_model.rsquared , final_model.aic
        print("Thus model accuracy is improved to",final_model.rsquared)
```
